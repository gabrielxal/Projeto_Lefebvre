## Análise exploratória nos dados de resíduos florestais




```python
dfForestryRatios = pd.read_excel('/content/drive/MyDrive/Analise Lefebvre/Dados/tabela 4 dados suplementares 1.xlsx')
```

```python
dfForestryRatios[['Collection factor\n\n[%]', 'Biochar yield\n(DM basis)\n[%]', 'Biochar organic carbon content (DM basis)\n[%]']] /= 100
```

```python
dfForestryRatios = dfForestryRatios.rename(columns = {'Dry bulk density\n\n[t m-3]': 'Dry Bulk density (t m-3)', 'Collection factor\n\n[%]': 'Collection factor',
       'Biochar yield\n(DM basis)\n[%]': 'Biochar yield (DM basis)',
       'Biochar organic carbon content (DM basis)\n[%]': 'Biochar organic carbon content (DM basis)'})
```

```python
dfForestryRatios
```

## Realizar cálculos e filtrar os resultados dos resíduos e da produção de biochar

```python

# Inicializar uma lista para armazenar os resultados
resultados = []
linhaRatios = dfForestryRatios.iloc[0]
# Iterar sobre os países
for pais in dfForestry['Area'].unique():
    # Filtrar as linhas correspondentes ao país atual
    df_producao_pais = dfForestry[dfForestry['Area'] == pais]

    # Inicializar a soma dos resíduos para o país atual
    soma_residuos_pais = 0
    soma_biochar = 0

    # Iterar sobre as culturas do país atual
    for index, row in df_producao_pais.iterrows():

      # Somar os resíduos considerando os coeficientes correspondentes do tipo de resíduo
      soma_biochar += row['Value'] * linhaRatios['Dry Bulk density (t m-3)'] * linhaRatios['Collection factor'] * linhaRatios['Biochar yield (DM basis)'] #* linhaRatios['Biochar organic carbon content (DM basis)']
      soma_residuos_pais += row['Value'] * linhaRatios['Dry Bulk density (t m-3)'] * linhaRatios['Collection factor'] # * linhaRatios['Biochar organic carbon content (DM basis)']


    # Armazenar o resultado para o país atual na lista
    resultados.append({'País': pais, 'Soma de Resíduos': soma_residuos_pais, 'Biochar Produzido': soma_biochar})

# Criar um DataFrame a partir da lista de resultados
df_resultados = pd.DataFrame(resultados)
```
