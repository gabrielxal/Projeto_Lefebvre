```python
dfRForestry = pd.read_excel('/content/drive/MyDrive/Analise Lefebvre/Dados quantificados/residues forestry.xlsx', usecols=['País', 'Soma de Resíduos', 'Biochar Produzido'])
```

```python
dfRWater = pd.read_excel('/content/drive/MyDrive/Analise Lefebvre/Dados quantificados/residues water.xlsx', usecols=['País', 'Soma de Resíduos','Biochar Produzido'])
```

```python
dfRAnimal = pd.read_excel('/content/drive/MyDrive/Analise Lefebvre/Dados quantificados/residues animal.xlsx', usecols=['País', 'Soma de Resíduos','Biochar Produzido'])
```

## Exploração Inicial e Cálculos

```python
dfRWater.info()
```

```python


df1 = dfRCrop
df2 = dfRAnimal
df3 = dfRForestry
df4 = dfRWater

# Merge dos DataFrames usando a coluna 'país' como chave de junção
merged_df = pd.merge(df1, df2, on='País', how='outer', suffixes=('_df1', '_df2'))
merged_df = pd.merge(merged_df, df3, on='País', how='outer')
merged_df = pd.merge(merged_df, df4, on='País', how='outer', suffixes=('_df3', '_df4'))

# Preenchendo valores nulos com 0
merged_df = merged_df.fillna(0)

# Somando as colunas
merged_df['Soma total de resíduos'] = merged_df['Soma de Resíduos_df1'] + merged_df['Soma de Resíduos_df2']+ merged_df['Soma de Resíduos_df3'] + merged_df['Soma de Resíduos_df4']
merged_df['Biochar total produzido'] = merged_df['Biochar Produzido_df1'] + merged_df['Biochar Produzido_df2']+ merged_df['Biochar Produzido_df3'] + merged_df['Biochar Produzido_df4']


```

```python

# Exibindo o DataFrame final
merged_df
```

```python
resultdf = pd.DataFrame({'Países': merged_df['País'],
    'Soma total de resíduos (Mg per year)': merged_df['Soma total de resíduos'],
    'Biochar total produzido (Mg per year)': merged_df['Biochar total produzido'],
    'YieldBC (%)': (merged_df['Biochar total produzido'] / merged_df['Soma total de resíduos']) * 100
})
```
